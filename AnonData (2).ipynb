If we are to better capture the current trend, we should restrict to only recent logins. Those in 2024, say.

```python
az_rec = arizona[arizona.date >= '2024-01-01']
az_rec_aggregated = az_rec.groupby('date')['count'].sum().reset_index()
```

```python
# Calculate a rolling average with a window of, say, 7 days
az_rec_aggregated['rolling_avg'] = az_rec_aggregated['count'].rolling(window=7).mean()

# Plot the original data and the rolling average
plt.figure(figsize=(12, 6))
sns.lineplot(data=az_rec_aggregated, x='date', y='count', label='Daily Count', alpha=0.5)
sns.lineplot(data=az_rec_aggregated, x='date', y='rolling_avg', label='7-Day Rolling Average')
plt.title('Daily Login Count and 7-Day Rolling Average in Arizona (2024)')
plt.xlabel('Date')
plt.ylabel('Number of Logins')
plt.legend()
plt.show()
```

```python
us_states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA",
             "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD",
             "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ",
             "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC",
             "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]

canada_provinces = ["AB", "BC", "MB", "ON", "QC"]

locations_of_interest = us_states + canada_provinces
```

```python
from scipy.stats import linregress

def calculate_slope_of_rolling_average(df_state):
    """
    Calculates the slope of the 7-day rolling average of the 'count' column
    for the last 30 days in a given DataFrame.

    Args:
        df_state: pandas DataFrame for a specific state or province
                  with 'date' and 'count' columns.

    Returns:
        The slope of the linear regression on the last 30 days of the
        rolling average, or None if there is not enough data.
    """
    # Ensure date column is datetime type and sort by date
    df_state = df_state.copy()
    df_state['date'] = pd.to_datetime(df_state['date'])
    df_state = df_state.sort_values(by='date')

    # Calculate the 7-day rolling average
    df_state['rolling_avg'] = df_state['count'].rolling(window=7).mean()

    # Select the last 30 days of the rolling average data, dropping NaNs
    last_days = df_state.tail(30).dropna(subset=['rolling_avg']).copy()

    if len(last_days) < 2:
        return None # Need at least two points for linear regression

    # Convert date to numerical representation (e.g., days since the first date)
    last_days['date_numeric'] = (last_days['date'] - last_days['date'].min()).dt.days

    # Check if there are at least two unique numerical date points
    if last_days['date_numeric'].nunique() < 2:
        return None

    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(last_days['date_numeric'], last_days['rolling_avg'])

    return slope

results = []
for location in locations_of_interest:
    df_location = df[df['state'] == location].copy() # Filter data for the current location
    slope = calculate_slope_of_rolling_average(df_location)
    results.append((location, slope))

display(results)
```

```python
slopes_df = pd.DataFrame(results, columns=['State/Province', 'Slope'])
slopes_df.sort_values(by='Slope', ascending=False).head()
```

```python
slopes_df.dropna().sort_values(by='Slope', ascending=False).tail()
```

## Summary
*   We calculated the slope of the 7-day rolling average of new logins for the last 30 days for 50 US states and 5 Canadian provinces.
*   Some locations have a 'Slope' value of `NaN`, indicating that there was insufficient data in the last 30 days to calculate a meaningful slope.

### Insights
*   The calculated slopes provide an indication of the recent trend in logins for each location. A positive slope suggests increasing an increasing number of customer logins, while a negative slope suggests logins are decreasing in number.
*   Finally, we sort the results by slope to identify locations with the steepest increases or decreases in logins.
*   The data suggests that Missouri and Arizona are experiencing significant growth in logins, so one may wish to focus efforts and money on retaining the new customers there and further growing the customer base.
*   We also note that Pennsylvania and North Carolina are experiencing a significant decline in logins (a particularly steep decline in PA's case), so one may wish either to shore up the customer base there, or alternatively to exit those markets.
